```python

from airfoil.airfoil import NACA4, NACA5

naca2412 = NACA4(designation="2412", chord_length=1.0, resolution=200)
naca2412.set_angle_of_attack(5.0)
naca2412.plot(title="NACA 2412 at 5 AoA")

naca2412.chord
naca2412.mean_camber_line
naca2412.upper_surface
naca2412.lower_surface
naca2412.thickness
naca2412.alpha

naca23012 = NACA5(designation="23012", chord_length=1.0, resolution=200)
naca23012.set_angle_of_attack(10.0)
naca23012.plot(title="NACA 23012 at 10 AoA")
```
